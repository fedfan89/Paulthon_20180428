```python
import numpy as np
import pylab as p
```

```python
data = np.array(np.random.rand(1000))
```

```python
np.histogram(data, bins=100)
```

```python
y, binEdges = np.histogram(data,bins=100)
```

```python
bincenters = .5*(binEdges[1:]+binEdges[:-1])
```

```python
p.plot(bincenters, y, '-')
```
